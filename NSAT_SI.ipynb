```python
from netCDF4 import Dataset
```

```python
nc_file = Dataset("Tair_WFDE5_CRU_197901_v2.1.nc", 'r')
```

```python
nc_file.variables
```

```python
target_x = 16.22  # Replace with your desired x-coordinate
target_y = 48.12  # Replace with your desired y-coordinate

# Find the indices corresponding to the specified coordinates
x_values = nc_file.variables['lon'][:]
y_values = nc_file.variables['lat'][:]
```

```python
nearest_x_index = (abs(x_values - target_x)).argmin()
nearest_y_index = (abs(y_values - target_y)).argmin()
```

```python
time_variable=nc_file.variables['time']
```

```python
nc_file.variables
```

```python
from datetime import datetime, timedelta
reference_date = datetime(1900, 1, 1)

# Calculate the dates and hours for each time step
dates_and_hours = [reference_date + timedelta(hours=float(hour)) for hour in time_variable[:]]

```

```python
dates_and_hours
```

```python
# Assuming you have a specific datetime value you want to find
target_datetime = datetime(1979, 1, 1, 12, 0)  # Replace with your target datetime

# Find the index for the target datetime in dates_and_hours
target_index = next((i for i, dt in enumerate(dates_and_hours) if dt == target_datetime), None)

if target_index is not None:
    print(f"Index for {target_datetime}: {target_index}")
else:
    print(f"{target_datetime} not found in dates_and_hours")
```
